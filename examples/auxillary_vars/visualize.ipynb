```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_lps = []
sp_lps = []
samples = []

data_folder = "row_trace_vs_sp_vs_sample_euler"

# Load the LPs from the pickle files
for i in range(6):
    my_lp = pickle.load(open(f"{data_folder}/lp_traj_{i}.pkl", 'rb'))
    sp_lp = pickle.load(open(f"{data_folder}/sp_lp_traj_{i}.pkl", 'rb'))
    sample = pickle.load(open(f"{data_folder}/sample_traj_{i}.pkl", 'rb'))

    my_lps.append(my_lp)
    sp_lps.append(sp_lp)
    samples.append(sample)
```

```python
for i in range(6):
    # Function to calculate Euclidean distance between upper and lower vectors
    def euclidean_distance(interval):
        return np.linalg.norm(interval.upper - interval.lower)

    # Calculate distances for each interval in the lists
    my_lp_distances = [euclidean_distance(interval) for interval in my_lps[i]]
    sp_lp_distances = [euclidean_distance(interval) for interval in sp_lps[i]]
    sample_distances = [euclidean_distance(interval) for interval in samples[i]]

    # Create x-axis values (interval indices)
    x_values = np.arange(len(my_lp))

    # Plot the distances
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, my_lp_distances, label='my_lp Distances', marker='o')
    plt.plot(x_values, sp_lp_distances, label='sp_lp Distances', marker='x')
    plt.plot(x_values, sample_distances, label='sample Distances', marker='s')
    plt.xlabel('Timestep')
    plt.ylabel('|upper - lower|')
    plt.title(f'Size of Interval Approximations over time ({i+1} aux vars)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for i in range(6):
    print(f"Final bound with {i+1} aux vars:")
    print(f"my_lp:\n{my_lps[i][-1][:2]}")
    print(f"sp_lp:\n{sp_lps[i][-1][:2]}")
```
